```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import random

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

def init():
    T_TOTAL = 6000
    T_UPPER_BOUND = 6500
    T_UNDER_BOUND = 2000
    PADDING = 500
    INTERVAL = 1
    INTERVAL_WHITE = 20
    INTERVAL_RED = 32
    INTERVAL_BLUE = 100
    
    # 잎
    z = [t for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL)]
    x = [(0.7 if t % 2 == 0 else 0.8) *(T_TOTAL-t)*math.cos(t/20) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING,INTERVAL)]
    y = [(0.7 if t % 2 == 0 else 0.8) *(T_TOTAL-t)*math.sin(t/20) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING,INTERVAL)]
    ax.scatter(x,y,z, c="green", marker="^")
    
    # 기둥
    z = [t for t in range(T_TOTAL-PADDING)]
    x = [300*math.cos(t/2) for t in range(T_TOTAL-PADDING)]
    y = [300*math.sin(t/2) for t in range(T_TOTAL-PADDING)]
    ax.scatter(x,y,z, c="#704214", marker="^")
    
    # 눈
    z = [t for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_WHITE)]
    x = [0.8*(T_TOTAL-t+10)*math.cos(t*13) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_WHITE)]
    y = [0.8*(T_TOTAL-t+10)*math.sin(t*13) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_WHITE)]
    ax.scatter(x,y,z, c="white", marker="p",s=40)
    
    # # 파란 장식
    z = [t for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_BLUE)]
    x = [0.33 *(T_TOTAL-t)*math.cos(t/20+10) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_BLUE)]
    y = [0.33 *(T_TOTAL-t)*math.sin(t/20+10) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_BLUE)]
    ax.scatter(x,y,z, c="blue", marker="o",s=10)
    
    # 빨간 장식
    z = [t for t in range(T_UNDER_BOUND+PADDING, T_UPPER_BOUND-PADDING, INTERVAL_RED)]
    x = [1 *(T_TOTAL-t)*math.cos(t/13) for t in range(T_UNDER_BOUND+PADDING, T_UPPER_BOUND-PADDING, INTERVAL_RED)]
    y = [1 *(T_TOTAL-t)*math.sin(t/13) for t in range(T_UNDER_BOUND+PADDING, T_UPPER_BOUND-PADDING, INTERVAL_RED)]
    ax.scatter(x,y,z, c="red", marker=".", s=130)
    
    # 별
    z = [t for t in range(T_UPPER_BOUND-5,T_UPPER_BOUND)]
    x = [0 for i in range(T_UPPER_BOUND-5,T_UPPER_BOUND)]
    y = [0 for i in range(T_UPPER_BOUND-5,T_UPPER_BOUND)]
    ax.scatter(x,y,z, c='yellow', marker="*",s=700,edgecolors='black',linewidths=0.7)
    
    plt.xlim(-6000,6000)
    plt.ylim(-6000,6000)

    return fig,

def animate(f):
    fig.clear()
    ax = fig.add_subplot(111, projection="3d")
    ax.set_facecolor('lightblue')
    ax.set_title("Merry Christmas", fontsize=16, fontweight='bold')

    # image_path = 'temp.jpeg'
    # image = plt.imread(image_path)
    # ax.imshow(image, extent=(0, 1, 0, 1), transform=ax.transAxes, aspect='auto', alpha=0.7)
    
    T_TOTAL = 6000
    T_UPPER_BOUND = 6500
    T_UNDER_BOUND = 2000
    PADDING = 500
    INTERVAL = 1
    INTERVAL_WHITE = 20
    INTERVAL_RED = 32
    INTERVAL_BLUE = 100
    
    # 잎
    z = [t for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL)]
    x = [(0.7 if t % 2 == 0 else 0.8) *(T_TOTAL-t)*math.cos(t/20) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING,INTERVAL)]
    y = [(0.7 if t % 2 == 0 else 0.8) *(T_TOTAL-t)*math.sin(t/20) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING,INTERVAL)]
    ax.scatter(x,y,z, c="green", marker="^")
    
    # 기둥
    z = [t for t in range(T_TOTAL-PADDING)]
    x = [350*math.cos(t/2) for t in range(T_TOTAL-PADDING)]
    y = [350*math.sin(t/2) for t in range(T_TOTAL-PADDING)]
    ax.scatter(x,y,z, c="#704214", marker="^")
    
    # 눈
    z = [t for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_WHITE)]
    x = [0.9*(T_TOTAL-t+10)*math.cos(t*13+f/10) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_WHITE)]
    y = [0.9*(T_TOTAL-t+10)*math.sin(t*13+f/10) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_WHITE)]
    ax.scatter(x,y,z, c="white", marker="p",s=40)
    
    # 파란 장식
    z = [t for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_BLUE)]
    x = [0.33 *(T_TOTAL-t)*math.cos(t/20+10+f/10) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_BLUE)]
    y = [0.33 *(T_TOTAL-t)*math.sin(t/20+10+f/10) for t in range(T_UNDER_BOUND, T_UPPER_BOUND-PADDING, INTERVAL_BLUE)]
    ax.scatter(x,y,z, c="blue", marker="o",s=10)
    
    # 빨간 장식
    z = [t for t in range(T_UNDER_BOUND+PADDING, T_UPPER_BOUND-PADDING, INTERVAL_RED)]
    x = [1 *(T_TOTAL-t)*math.cos(t/13+f/10) for t in range(T_UNDER_BOUND+PADDING, T_UPPER_BOUND-PADDING, INTERVAL_RED)]
    y = [1 *(T_TOTAL-t)*math.sin(t/13+f/10) for t in range(T_UNDER_BOUND+PADDING, T_UPPER_BOUND-PADDING, INTERVAL_RED)]
    ax.scatter(x,y,z, c="red", marker=".", s=130)
    
    # 별
    z = [t for t in range(T_UPPER_BOUND-5,T_UPPER_BOUND)]
    x = [0 for i in range(T_UPPER_BOUND-5,T_UPPER_BOUND)]
    y = [0 for i in range(T_UPPER_BOUND-5,T_UPPER_BOUND)]
    ax.scatter(x,y,z, c='yellow', marker="*",s=500,edgecolors='black',linewidths=0.7)

    for _ in range(10):
        z = random_number = random.randint(0, 6000)
        x = random_number = random.randint(-6000, 6000)
        y = random_number = random.randint(-6000, 6000)
        ax.scatter(x,y,z, c="yellow", marker="|", s=70)
        
    for _ in range(20):
        z = random_number = random.randint(0, 6000)
        x = random_number = random.randint(-6000, 6000)
        y = random_number = random.randint(-6000, 6000)
        ax.scatter(x,y,z, c="white", marker="o", s=50, edgecolors='black', linewidths=0.3)
        
    # 3D 그래프에 레전드 추가
    ax.text2D(0.8, 0.02, f"Frame {f}", transform=ax.transAxes, fontsize=10, va='bottom', ha='right')
    
    plt.xlim(-6000,6000)
    plt.ylim(-6000,6000)

    return fig,

ani = animation.FuncAnimation(fig, animate, init_func=init,
                              frames=90, interval=70, blit=True)

from IPython.display import HTML
HTML(ani.to_html5_video())
```
